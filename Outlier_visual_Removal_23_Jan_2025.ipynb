```python
#def remove_outliers(df, column):
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_filtered = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
    #return df_filtered
```

```python
Q1 = df_filtered['Fare'].quantile(0.25)
Q3 = df_filtered['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_filtered = df_filtered[(df_filtered['Fare'] >= lower_bound) & (df_filtered['Fare'] <= upper_bound)]
```

```python
x = df_filtered.iloc[:,1:8]
y = df_filtered.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=345)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
ypredict = model.predict(x_test)
```
