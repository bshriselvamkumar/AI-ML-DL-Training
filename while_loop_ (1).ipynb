Emulating a do-while loop in Python
There are two scenarios in which a loop terminates:

The loop condition is no longer true/false (depending on the type of loop).
A break statement is executed from within the code in the loop body.


```python
i = 10

while True:  #while boolean value
    print(i)
    i = i + 10
    if(i > 50):
      break
```

```python
i=0
while(i<10): # condition
  print(i)
  i=i+3
```

```python
i = 10

while False:
    print(i)
    i = i - 1
    if(i < 3):
        break
```

```python
import random

magic_number = random.randint(1,20)

attempts = 0

while attempts < 5:
	print("Guess a number between 1 and 20:")

	guess = int(input())

	attempts = attempts + 1

	if guess == magic_number: # while with if condition external terminator break stmt needed
		break

print("You have guessed the magic number!")
```
